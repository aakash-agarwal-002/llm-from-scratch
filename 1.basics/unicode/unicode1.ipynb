### Unicode
Unicode is a text encoding standard that maps characters to integer code points.
As of Unicode 17.0 (released in September 2025), the standard defines 159,801 characters across 172 scripts.

#### ord and chr
- ord(): function to convert a single Unicode character into its integer representation. 
- chr(): function converts an integer Unicode code point into a string with the corresponding character.

```python
ord('😂')
```

```python
chr(128514)
```

#### Null character

```python
chr(0) # prints null character in string representation (__repr__()) form but it is a valid character
```

```python
print(chr(0)) # prints nothing in printed representation, but it is a valid character
print(len(chr(0))) # prints 1, because it is a single character
```

```python
"this is a test" + chr(0) + "string" # concatenates the string with the null character in between
```

```python
print("this is a test" + chr(0) + "string")
```
